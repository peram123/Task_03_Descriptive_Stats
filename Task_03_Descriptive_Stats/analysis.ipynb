```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load your dataset
df = pd.read_csv('/workspaces/codespaces-blank/2024_fb_ads_president_scored_anon.csv')

# -- Histogram for Estimated Spend --
plt.figure(figsize=(8,4))
sns.histplot(df['estimated_spend'].astype(float), bins=50, kde=True)
plt.title('Distribution of Estimated Spend')
plt.xlabel('Estimated Spend')
plt.ylabel('Count')
plt.show()

# -- Boxplot for Estimated Impressions --
plt.figure(figsize=(8,4))
sns.boxplot(x=df['estimated_impressions'].astype(float))
plt.title('Boxplot of Estimated Impressions')
plt.xlabel('Estimated Impressions')
plt.show()

# -- Bar chart for Top 10 Pages by Number of Ads --
top_pages = df['page_id'].value_counts().head(10)
plt.figure(figsize=(10,4))
sns.barplot(x=top_pages.index, y=top_pages.values, palette='viridis')
plt.title('Top 10 Pages by Number of Ads')
plt.xlabel('Page ID')
plt.ylabel('Number of Ads')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# -- Bar chart for Currency Frequency --
currency_counts = df['currency'].value_counts()
plt.figure(figsize=(6,3))
sns.barplot(x=currency_counts.index, y=currency_counts.values, palette='Set2')
plt.title('Currency Usage in Ads')
plt.xlabel('Currency')
plt.ylabel('Count')
plt.tight_layout()
plt.show()

# -- Proportion of Issue/Attack Ads --
for col in ['issue_msg_type_illuminating', 'attack_msg_type_illuminating']:
    val_counts = df[col].value_counts(normalize=True)
    plt.figure(figsize=(3,3))
    sns.barplot(x=val_counts.index, y=val_counts.values)
    plt.title(f'Proportion of {col}')
    plt.ylabel('Proportion')
    plt.xlabel('Flag Value')
    plt.tight_layout()
    plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# ---- Change this filename for each dataset ----
filename = '/workspaces/codespaces-blank/2024_fb_posts_president_scored_anon.csv'
# filename = '2024_fb_posts_president_scored_anon.csv'
# filename = '2024_tw_posts_president_scored_anon.csv'

df = pd.read_csv(filename)
print(f"{filename}: shape = {df.shape}")
display(df.head())

# Histogram for numeric columns (if present)
numeric_cols = ['estimated_spend', 'estimated_impressions', 'estimated_audience_size']
for col in numeric_cols:
    if col in df.columns:
        plt.figure(figsize=(8,4))
        sns.histplot(df[col].astype(float), bins=50, kde=True)
        plt.title(f'Distribution of {col}')
        plt.xlabel(col)
        plt.ylabel('Count')
        plt.show()

# Boxplot for numeric columns
for col in numeric_cols:
    if col in df.columns:
        plt.figure(figsize=(8,4))
        sns.boxplot(x=df[col].astype(float))
        plt.title(f'Boxplot of {col}')
        plt.xlabel(col)
        plt.show()

# Top 10 pages by number of ads/posts
page_col = None
for possible_col in ['page_id', 'twitter_page_id', 'fb_page_id']:
    if possible_col in df.columns:
        page_col = possible_col
        break
if page_col:
    top_pages = df[page_col].value_counts().head(10)
    plt.figure(figsize=(10,4))
    sns.barplot(x=top_pages.index.astype(str), y=top_pages.values, palette='viridis')
    plt.title('Top 10 Pages by Count')
    plt.xlabel(page_col)
    plt.ylabel('Count')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()

# Bar chart for most common currency
if 'currency' in df.columns:
    currency_counts = df['currency'].value_counts()
    plt.figure(figsize=(6,3))
    sns.barplot(x=currency_counts.index, y=currency_counts.values, palette='Set2')
    plt.title('Currency Usage')
    plt.xlabel('Currency')
    plt.ylabel('Count')
    plt.tight_layout()
    plt.show()

# Proportion of key binary flags
binary_flag_cols = [col for col in df.columns if col.endswith('_illuminating') or col.endswith('_type_illuminating')]
for col in binary_flag_cols:
    if col in df.columns:
        val_counts = df[col].value_counts(normalize=True)
        plt.figure(figsize=(3,3))
        sns.barplot(x=val_counts.index.astype(str), y=val_counts.values)
        plt.title(f'Proportion of {col}')
        plt.ylabel('Proportion')
        plt.xlabel('Value')
        plt.tight_layout()
        plt.show()
```
